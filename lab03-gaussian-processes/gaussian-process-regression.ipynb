# Gaussian Process Regression

## Aims
* To sample from a Gaussian process prior distribution.
* To implement Gaussian process inference for regression.
* To use the above to observe samples from a Gaussian process posterior distribution.
* To evaluate how different hyperparameter settings impact model quality.

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Fira Sans'
plt.rcParams['figure.figsize'] = (16.0, 8.0)
```

```python
def f(x):
    return np.sin(np.exp(0.03 * x))
```

```python
x = np.random.uniform(low=-20, high=80, size=(200,))
t = f(x)
```

```python
fig, ax = plt.subplots()
fig.suptitle('Sampled points', fontsize=18)
ax.plot(np.arange(-20, 80), f(np.arange(-20, 80)), color='red')
ax.scatter(x, t)
plt.show()
```
